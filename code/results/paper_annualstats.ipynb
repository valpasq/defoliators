```python
%matplotlib inline

import numpy as np
import rasterio
from osgeo import gdal
import matplotlib.pyplot as plt
from cycler import cycler
import matplotlib.ticker as tkr
import matplotlib.patches as mpatches
import rasterio
import seaborn as sns
from __future__ import division
import pandas as pd
import seaborn
```

```python
# set style preferences
plt.style.use('dark_background') 
#plt.style.use('ggplot')
plt.rcParams['axes.prop_cycle'] = cycler(color='bgrcmyk')
plt.rcParams['grid.color'] = '0.5'
plt.rcParams['grid.linestyle'] = ':'
plt.rcParams['grid.linewidth'] = '1'
plt.rcParams['font.size'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 12
```

```python
# read in 2015 assessment
with rasterio.open('../2015_analysis/final_products/2015_0515_meanresiduals_forestmask.tif') as image:
    cond_2015 = image.read(1)
print(cond_2015.shape)
cond_2015[cond_2015 == -9999] = 9999 # convert nodata to positive value

# read in 2016 assessment
with rasterio.open('../2016_analysis/final_products/2016_0515_meanresiduals_forestmask.tif') as image:
    cond_2016 = image.read(1)
print(cond_2016.shape)
cond_2016[cond_2016 == -9999] = 9999 # convert nodata to positive value

# read in 2017 assessment
with rasterio.open('../2017_analysis/final_products/2017_0515_meanresiduals_forestmask.tif') as image:
    cond_2017 = image.read(1)
print(cond_2017.shape)
cond_2017[cond_2017 == -9999] = 9999 # convert nodata to positive value
```

```python
print('2015 Assessment')
print('------')

print('Area (km2):')
cond_2015_1 = cond_2015[cond_2015 < -1].size * 900/(1000*1000)
print(cond_2015_1)
cond_2015_2 = cond_2015[cond_2015 < -2].size * 900/(1000*1000)
print(cond_2015_2)
cond_2015_3 = cond_2015[cond_2015 < -3].size * 900/(1000*1000)
print(cond_2015_3)
cond_2015_4 = cond_2015[cond_2015 < -4].size * 900/(1000*1000)
print(cond_2015_4)

print('------')
print('Percent of Forested Area:')
print(((cond_2015[cond_2015 < -1].size * 900/(1000*1000)) / cond_2015[cond_2015 != 9999].size) * 100)
print(((cond_2015[cond_2015 < -2].size * 900/(1000*1000)) / cond_2015[cond_2015 != 9999].size) * 100)
print(((cond_2015[cond_2015 < -3].size * 900/(1000*1000)) / cond_2015[cond_2015 != 9999].size) * 100)
print(((cond_2015[cond_2015 < -4].size * 900/(1000*1000)) / cond_2015[cond_2015 != 9999].size) * 100)
```

```python
print('2016 Assessment')
print('------')

print('Area (km2):')
cond_2016_1 = cond_2016[cond_2016 < -1].size * 900/(1000*1000)
print(cond_2016_1)
cond_2016_2 = cond_2016[cond_2016 < -2].size * 900/(1000*1000)
print(cond_2016_2)
cond_2016_3 = cond_2016[cond_2016 < -3].size * 900/(1000*1000)
print(cond_2016_3)
cond_2016_4 = cond_2016[cond_2016 < -4].size * 900/(1000*1000)
print(cond_2016_4)

print('------')
print('Percent of Forested Area:')
print(((cond_2016[cond_2016 < -1].size * 900/(1000*1000)) / cond_2016[cond_2016 != 9999].size) * 100)
print(((cond_2016[cond_2016 < -2].size * 900/(1000*1000)) / cond_2016[cond_2016 != 9999].size) * 100)
print(((cond_2016[cond_2016 < -3].size * 900/(1000*1000)) / cond_2016[cond_2016 != 9999].size) * 100)
print(((cond_2016[cond_2016 < -4].size * 900/(1000*1000)) / cond_2016[cond_2016 != 9999].size) * 100)
```

```python
print('2017 Assessment')
print('------')

print('Area (km2):')
cond_2017_1 = cond_2017[cond_2017 < -1].size * 900/(1000*1000)
print(cond_2017_1)
cond_2017_2 = cond_2017[cond_2017 < -2].size * 900/(1000*1000)
print(cond_2017_2)
cond_2017_3 = cond_2017[cond_2017 < -3].size * 900/(1000*1000)
print(cond_2017_3)
cond_2017_4 = cond_2017[cond_2017 < -4].size * 900/(1000*1000)
print(cond_2017_4)

print('------')
print('Percent of Forested Area:')
print(((cond_2017[cond_2017 < -1].size * 900/(1000*1000)) / cond_2017[cond_2017 != 9999].size) * 100)
print(((cond_2017[cond_2017 < -2].size * 900/(1000*1000)) / cond_2017[cond_2017 != 9999].size) * 100)
print(((cond_2017[cond_2017 < -3].size * 900/(1000*1000)) / cond_2017[cond_2017 != 9999].size) * 100)
print(((cond_2017[cond_2017 < -4].size * 900/(1000*1000)) / cond_2017[cond_2017 != 9999].size) * 100)
```

```python
colors = ["#feed26", "#fe9929", "#e31a1c", "#b50006"]
```

```python
fig = plt.figure(figsize=(6,10))
ax1 = plt.axes(frameon=False)
plt.bar([0.8, 1.8, 2.8, 3.8], [cond_2015_1, cond_2015_2, cond_2015_3, cond_2015_4], width=0.2, label='2015')
plt.bar([1, 2, 3, 4], [cond_2016_1, cond_2016_2, cond_2016_3, cond_2016_4], width=0.2, label='2016')
plt.bar([1.2, 2.2, 3.2, 4.2], [cond_2017_1, cond_2017_2, cond_2017_3, cond_2017_4], width=0.2, label='2017')
plt.ylabel('Area (km2)')
plt.ylim((0, 5000))
plt.yticks(size=12)
plt.xlabel('Condition assessment')
plt.xticks([1, 2, 3, 4], ('Slight change', 'Moderate change', 'Large change', 'Very large change'), size=10)

plt.legend()
plt.savefig('../yeartoyearcomparison.png', transparent=True, dpi=300)
plt.show()
```

```python
fig = plt.figure(figsize=(6,10))
ax1 = plt.axes(frameon=False)

plt.bar([1, 1, 1, 1], [cond_2015_1, cond_2015_2, cond_2015_3, cond_2015_4], color=colors, width=0.2)
plt.bar([1.5, 1.5, 1.5, 1.5], [cond_2016_1, cond_2016_2, cond_2016_3, cond_2016_4], color=colors, width=0.2)
plt.bar([2, 2, 2, 2], [cond_2017_1, cond_2017_2, cond_2017_3, cond_2017_4], color=colors, width=0.2)
plt.ylabel('Area (km2)')
plt.ylim((0, 5000))
plt.yticks(size=14)
plt.xticks([1, 1.5, 2], ('2015', '2016', '2017'), size=16)

cond_1 = mpatches.Patch(color=colors[0], label='Slight change')
cond_2 = mpatches.Patch(color=colors[1], label='Moderate change',)
cond_3 = mpatches.Patch(color=colors[2], label='Large change')
cond_4 = mpatches.Patch(color=colors[3], label='Very large change')
leg = plt.legend(handles=[cond_1, cond_2, cond_3, cond_4], loc='best')

plt.legend()
plt.tight_layout()
plt.savefig('../yeartoyearcomparison_stacked.png', transparent=True, dpi=300)
plt.show()
```
